# The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

here using paired sample t test beacuse have to check whether there is any increase in sales after
stepping into digital marketing.These both condion are dependented.


H0:there is no increase in sales after stepping into digital marketing.
Ha:there is increase in sales after stepping into digital marketing.   

```python
data=pd.read_csv('Sales_add.csv')
#reading Sales_before_digital_add
Sales_before_digital_add=data['Sales_before_digital_add(in $)']
Sales_before_digital_add
```

```python
#reading Sales_After_digital_add
Sales_After_digital_add=data['Sales_After_digital_add(in $)']
Sales_After_digital_add
```

```python
from scipy.stats import ttest_rel
ttest_rel(Sales_before_digital_add,Sales_After_digital_add)
```

comparing p value with significance level(0.05) 
pvalue is greater so reject the null hypothesis
ie,there is increase in sales after stepping into digital marketing. 

# The company needs to check whether there is any dependency between the features “Region” and “Manager”


H0:Region and Manager are independent 
Ha:Region and Manager are dependent    

```python
data=pd.read_csv('Sales_add.csv')
crosstab=pd.crosstab(data['Region'],data['Manager'],normalize='index')
crosstab
```

```python
from scipy import stats
stats.chi2_contingency(crosstab)
```

first value is the Chi-square value=0.509
second valueis p-value =0.97
third value the degrees of freedom=4


```python
level of significance is 0.05 is lessthan that of p value , so so null hypothesis is accepted.
Region and Manager are independent, ie no relation between Region and Manager for the sales
```
